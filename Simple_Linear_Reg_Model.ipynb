#***LINEAR REGRESSION MODEL [SALARY-PREDICTION]***

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
ds=pd.read_csv("Salary_dataset.csv",index_col=0)
```

```python
ds
```

```python
#Equation for the Best Fit Line
#h_theta = theta0 + (theta1*x)
```

```python
#Equation of Gradient Descent -> Theta_j = Theta_j - alpha*(delta (J_theta)/delta(theta_j))
#-> 'alpha' is the learning rate (Generally Taken 0.001)
def gradient_descent(x,y,theta0,theta1,alpha,m):
  temp0 = 0
  temp1 = 0
  for i in range (m):
    temp0 += (theta0 + (theta1*x[i])-y[i])
    temp1 += ((theta0 + (theta1*x[i])-y[i])*x[i])

  theta0 = theta0 - (alpha*(temp0/m))
  theta1 = theta1 - (alpha*(temp1/m))
  return theta0,theta1
```

```python
def costfunction(theta0,theta1,x,y):
  m = len(ds.YearsExperience)
  sq_error = 0
  alpha = 0.001
  #Repeating Until Convergence [Manually, Iterating 1000 times]
  for i in range (1000):
    theta0,theta1 = gradient_descent(x,y,theta0,theta1,alpha,m)
  for i in range (m):
    h_thetax = theta0+theta1*(x[i])
    sq_error += (h_thetax-y[i])**2
  J_theta = (1/(2*m))*sq_error
  # print(sq_error)
  return theta0,theta1,J_theta
```

```python
theta1 = 0
theta0 = 0
x=ds.YearsExperience
y=ds.Salary
theta0,theta1,J_theta = costfunction(theta0,theta1,x,y)
print(f"Mean Squared Error For Theta_1 ({theta1}) and Theta_0 ({theta0}) Is {J_theta}")
```

```python
plt.scatter(x,y)
plt.plot(x,theta0+(theta1*x),'r', marker= 'o', mfc = 'g', mec = 'g')
plt.xlabel("Years Experience")
plt.ylabel("Predicted Salary")
plt.show()
```
